# 2. What is Data Engineering?
* The field of DE concern itself with the mechanics for the flow and access of data. **Its goal is to make quality data available for fact-finding and data-driven decision making.**
* *The world of data has grown to include wide-ranging sources, structures, and types of data.*
* The field of Data Engineering involves:
    * `Collecting source data:`Extract, integrating, and organizing data from disparate sources
        * Data acquisiton from multiple sources
        * Data architecture for storing source data
    * `Processing data:` Cleaning, transforming, and prepareing data to make it usable
        * Distributed systems for processing data
        * Pipelines for extracting, transforming, and loading data
        * Solutions for safeguarding quality, privacy, and security of data
        * Performance optimization
        * Adherence to compliance guidlines
    * `Storing data:` Storing data for reliability and easy availability of data
        * Data stores for storage of processed data
        * Scalable systems
        * Ensuring data privacy, security, compliance, monitoring, backup, and recovery
    * `Making data available to users securely:` Making data available to users securely
        * APIs, services, and progams for retrieving data for end-users
        * User access through interfaces and dashboards
        * Checks and balances to ensure data security
* `Consider Data Engineering as a team sport:`
    * No one is expected to know everything
    * For example, to architect any data management syste, be it for collating source data or storing processed analysis-ready data, we need to have the skills of an architect. To ensure data stores are available and optimized for use, you need to have expertise in databases.
    * Similarly, proficiency in database tools, progamming languages, and distributed systems all come under data engineering, but they may require different skill sets
    * Not all teams and organizations need to set up and end-to-end data engineering practice. These are tools, applications, and solutions available in the market, both on-premise and cloud-based, that can be evaluated for individual needs.
    * `Summarize:` More than any other data profession, ***DE is about the tools and technologies involved in data manipulation***. But it is also about understanding the complexities of data and how it is ultimately leveraged for fact-finding and decision-making.

# 2. Types of Data
* What is data?
    * Data is unorganized information that is processed to make it meaningful
    * Data comprises of facts, observations, perceptions, number, characters, symbols, and images that can be interpreted to derive meaning.
* Data can be categorized is by its structure:
    * `Structured:`
        * Has a well-defined structure
        * Can be stored in well-defined schemas
        * Can be represented in a tabular manner with rows and columns
        * Structured data is objective facts and numbers that can be `collected`, `exported`, `stored`, and `organized` in typical databases
        * `Sources:` SQL Databases, Online Transaction Processing, Spreadsheets, Online forms, Sensores GPS and RFID, Network and Web server logs
    * `Semi-structured:`
        * Has some organizational properities but lacks a fixed or rigid scheme
        * Cannot be stored in the form of rows and columns as in databases
        * Contains tags and elements, or metadata, which is used to group data and organize it in a hierarchy
        * `Sources:` E-mails, XML and other markup languages, Binary executables, TCP/IP packets, Zipped files, Intergration of data
        * XML and JSON allow users to `Define Tags`, `Attributes`, `To store data` semi-structured data
    * `Unstructured:`
        * Does not have an easily identifiable structure
        * Cannot be organized in a mainstream relational database in the form of rows and columns
        * Does not follow any particular format, sequence, semantics, or rules
        * `Sources:` Web pages, Social media feeds, Images in varied file formats, Videos and Audio files, Documents and PDF files, PowerPoint presentations, Media logs, and Surveys
        * Data can be stored in files document for manual analysis or in NoSQL databases that have the own analysis tools for examining this type of data
### Summarize
* Structured data is data that is well organized in format that can be stored in databases and lends itself to standard data analysis methods and tools
* Semi-structured data is data that is somewhat organized and relies on meta tags for grouping and hierarchy
* Unstructured data is data that is not conventionally organized in the form of rows and columns in a particular format
# 3. Understanding different types of file formats
* `Standard file formats:`
    * `Delimited text file formats, or .CSV:`
        * Files used to store data as text
        * Each value is separated by a delimiter
        * `Delimiter` - A sequence of one or more characters for specifying the boundary between independent entities or values.
        * *Any character can be used to separate the values*, but most common delimiters are the *comma, tab, colon, vertical bar, and space*. `Comma-separated values (or CSVs)` and `Tab-separated values (or TSVs)` are the most commonly used files types in this category
        * 
    * `Microfoft Excel Open .XML Spreadsheet, or .XLSX:`
        * Is a Microsoft Excel Open XML file format that falls under the spreadsheet file format. It is an XML-based file format created by Microsoft
        * Open file format, accessible to most other applications
        * Can use and save all functions available in excel
        * Is a secure file format as it cannot save malicious code
    * `Extensible Markup Language, or .XML:`
        * Is a markup language with set rules for encoding data
        * Readable by both humans and machines
        * Self-descriptive language
        * Similar to .HTML in som respects
        * Does not use predefined tags like .HTML does
        * Platform independent
        * Programming language independent
        * Makes it simplier to share data between systems
    * `Portable Document Format, or .PDF:`
        * Is a file format developed by Adobe to present documents independent of application software, hardware, and operating systems
        * Can be viewd the same way on any device
        * Is frequently used in legal and financial documents
        * Can also be used to fill in data for forms 
    * `JavaScript Object Notation, or .JSON:`
        * Is a text-based standard designed for transmitting structured data over the web
        * language-independent data format
        * Can be read in any programming language
        * Easy to use
        * Compatible with a wide range of browsers
        * Considered as one of the best tools for sharing data 
# 4. Sources of Data
* `Common sources of data:`
    * `Relational databases:`
        * Typically, organizations have internal applications to support them in managing their day to day business activities, customer transactions, human resources activities, and their workflows
        * These system use relational databases such as `SQL Server`, `Oracle`, `MySQL`, and `IBM DB2`, to store in a structured way
        * Store structured data that can be leveraged for analysis
        * For example, data from a retail transactions system can be used to analyze sales in diffferent regions, and data from a customer relationship management system can be used for making sales projections
    * `Flat files and XML Datasets:`
        * `Flat file:`
            * ` Store data in plain text format, with one record or row per line, and each value saparated by a delimiters such as commas, semi-colons, or tabs
            * Data in a flat files maps to a single table, unlike relational databases that contain multiple tables
            * One of the most common flat file-format is CSV in which values are separated by commas
        * `Spreadsheet:`
            * Special type of flat files
            * Organize data in a tabular format
            * Can contain multiple worksheets
            * .XLS or .XLSX are common spreadsheet formats
            * Other formats include Google Sheets, Apple Numbers, and LibreOffice Calc 
        * `XML files:`
            *  Contain data values that are identified or marked up using tags
            *  Can support complex data structures
            *  Common uses include online surveys, bank statements, and other unstructured data sets
            *  
    * `APIs(Application Progam Interfaces) and Web Services:`
        * APIs and Web services listen for incoming request, which can be in the form of web requests from users or network requests from applications, and return data in plain text, XML, HTML, JSON, or media files.
        * For example, the use of Twitter and Facebook APIs to source data from tweets and post for performing task such as opinion mining or sentiment analysis - which is to summarize amount of appreciation and criticsm on a given subject, such as policies of a government, a product, a service, or customer satisfaction in general 
    * `Web Scraping:`
        * Extract relevant data from unstructured sources
        * Also known as Sreen scrapping, Web harvesting, and Web data extraction
        * Downloads specific data based on defined parameters
        * Can extract text, contact information, images, videos, product items, and more,...
        * `Popular uses:`
            * Providing price comparisions by collecting product details from retailer, manufacturers, and eCommerce websites
            * Generating sales lead through public data sources
            * Extracting data from post and author on various forums and communities
            * Collecting training and testing datasets for machine learning models
        * `Popular Web scraping tools:`
            * BeautifulSoup
            * Scrapy
            * Pandas
            * Selenium
    * `Data Streams and Feeds:`
        * Aggregeting streams of data flowing from instrument, IOT devices and applications, GPS data from car, computer programs, websites, and social media posts
        * `Some of the data streams and ways in wich they can be leveraged include:`
            * Stock and market tickets for financial trading
            * retail transaction stream for predicting demand and supply chain management
            * Surveillance and video feeds for thret detection
            * Social media feeds for sentiment analysis
            * Sensor data feeds for monitoring industrial or farming machinery
            * Web click feeds for monitoring web performance and improving design
            * Real-time flight events for rebooking and rescheduling
        * `Popular technologies used to process data strems include:`
            * Kafka
            * Apache Spark
        * `RSS(or Really Simple Syndication) feeds:`
            * Capturing updated data from online forums and news sites where data is refreshed on an ongoing basis
# 5. Languages for Data Professionals
* `SQL:`
    * Is a querying language designed for accessing and manipulating information from, mostly, though not exclusively, relational databases
    * `Using SQL:`
        * Insert, update, and delete records in a database
        * Create new databases, tables, and views
        * Write stored procedures
    * `Advantages:`
        * SQL is portable and platform independent
        * Can be used for querying data in a wide variety of databases and data repositories
        * Has a simple syntax that is similar to the English language
        * Its syntax allows developers to write programs with fewer lines of code using basic efficiently
        * Runs on an interpreter system
* `Python:`
    * Is a widely-used open-source, general-purppose, high-level programming language
    * Its syntax allows programmers to express their concepts in fewer lines of code
    * An ideal tool for beginning programmers because of its focus on simplicity and readability
    * Great for performing hig-computatioal tasks in large voulumes of data
    * Has in-built functions for frequently used concepts
    * Support multiple programming paradigms-object-oriented, imperative, functional, and procedural
    * `Advantages:`
        * Easy to learn
        * Open-source
        * Can be ported to multiple platforms
        * Has widespread community support
        * Provides open-source libraries for data manipulation, data visualization, statistics, mathematics
    * `Libraries:`
        * `Pandas` for data cleaning and analysis
        * `Numpy` and `Scipy`, for statistical analysis
        * `Beautifulsoup` and `Scrapy` for web scraping
        * `Matplotlib` and Seaborn to visually represent data in the form of bar graphs, histogram, and pie-charts
        * Opency for image processing
* `R:`
    * Is an open-source programming language and environment for data analysis, data visualization, machine learning, and statistics
    * `Widely used for:`
        * Developing statistical sofware
        * Performing data analytics
        * Creating compelling visualizations
    * `Key benefits:`
        * Open-source
        * Platform-independent
        * Can be paired with many programming languages
