    ## 1.2. Key Players in the Data Ecosystem
    * To get value from data, you need a vast number of skillsets and people playing different roles
    1. `Data Engineer:`
        * Are the people who develop and maintain data architectures and make data available for business operation and analysis
        * DE work within the ecosystem to extract, integrate, and organize data from disparate sources: clean, tranform, and prepare data; design, store and manage data in data repositories
        * DE enable data to be accessible in formats and systems that various business applications, as well as stakeholders like Data Analyst (DA) and Data Scientists (DS) , can utilize

        * DS analyze data for actionable insights and build Machine Learning and Deep Learning models that train on past data to create predictive models

    4. `Business Analyst and BI Analyst:`
        * BA leverage the work of DA and DS to look at posible implications for their business and the actions they need to take recommend.
        * BI do the same, except their focus is on market forces and external influences that shape their business. They provide business intelligence solution by organizing and monitoring data on different funtion and exploring that data to extract insights and acationables that improve business performance
        ### Summarize
       * DE converts raw data into usable data
       * DA uses this data to generate insights
       * DS use Data Analytics and DE to predict the future using data from the past
       * BA and BI use these insights and predictions to drive decisions that benefit and grow their business
# 2. What is Data Engineering?
* The field of DE concern itself with the mechanics for the flow and access of data. **Its goal is to make quality data available for fact-finding and data-driven decision making.**
* *The world of data has grown to include wide-ranging sources, structures, and types of data.*
* The field of Data Engineering involves:
    * `Collecting source data:`Extract, integrating, and organizing data from disparate sources
        * Data acquisiton from multiple sources
        * Data architecture for storing source data
    * `Processing data:` Cleaning, transforming, and prepareing data to make it usable
        * Distributed systems for processing data
        * Pipelines for extracting, transforming, and loading data
        * Solutions for safeguarding quality, privacy, and security of data
        * Performance optimization
        * Adherence to compliance guidlines
    * `Storing data:` Storing data for reliability and easy availability of data
        * Data stores for storage of processed data
        * Scalable systems
        * Ensuring data privacy, security, compliance, monitoring, backup, and recovery
    * `Making data available to users securely:` Making data available to users securely
        * APIs, services, and progams for retrieving data for end-users
        * User access through interfaces and dashboards
        * Checks and balances to ensure data security
* `Consider Data Engineering as a team sport:`
    * No one is expected to know everything
    * For example, to architect any data management syste, be it for collating source data or storing processed analysis-ready data, we need to have the skills of an architect. To ensure data stores are available and optimized for use, you need to have expertise in databases.
    * Similarly, proficiency in database tools, progamming languages, and distributed systems all come under data engineering, but they may require different skill sets
    * Not all teams and organizations need to set up and end-to-end data engineering practice. These are tools, applications, and solutions available in the market, both on-premise and cloud-based, that can be evaluated for individual needs.
    * `Summarize:` More than any other data profession, ***DE is about the tools and technologies involved in data manipulation***. But it is also about understanding the complexities of data and how it is ultimately leveraged for fact-finding and decision-making.

# 1. Overview of Data Engineering Ecosystem
* A Data Engineering's ecosystem includes the infrastructure, tools, frameworks and processes for:
    * Extracting data from disparare sources
    * Architecting and managing data pipelines for transformation, intergration, and storage of data
    * Architecting and managing data repositories
    * Automating and optimizing workflows and flows of data between systems
    * Developing applications needed through the data engineering workflow
* Data can be categorized as:
    * `Structured:` Data that follows a rigid format and **can be organized into rows and columns**
    * `Semi-structured:` **Mix of data** that has consistent characteristics and data that **does not conform to a rigid structure**
    * `Unstructured:` Data that is complex and mostly qualitative information that **can not be stuctured into rows and columns** for example, photos, videos, text files,...
* Data also comes in a ide-ranging viety of file formats being collected from a viety of data sources: Relational databases. non-relational database, APIs, web services, data streams, social platforms, sensor devices,...
* `Data repositories:`
    * `Transactional or Online Transaction Processing (or OTL) systems:`
        * Designed to store high-volume day-to-day operational data. For example. online banking transactions, airline bookings
        * Typically relational, but can also be non-relational
    * `Analytical or Online Analytical Processing (OLAP) Systems:`
        * Optimizied for conducting complex data analytics
        * Include relational and non-relational databases, data warehouses, data marts, data lakes, and big data stores
* `Data Intergration:`
    * Collated -> Processed -> cleansed -> Integrated -> Users
    * Combine data from disparate sources into a unified view, accessed by users to query and manipulate the data
* `Data pipeline:`
    
    * A set of tools and processes that cover the entire journey of data from source to destination systems
* `Language:`
    * Laguages available in the Data Analyst Ecosystem:
        * `Query languages:` For example, SQL for quying and manipulating data
        * `Programming languages:` For example, Python for developing data applications
        * `Shell and Scripting languages:` For repetitive operational tasks
* `Business Intelligence (BI) and Reporting Tools:`
    * Collecting data from multiple data sources and present them in a visual format, such as interactive dashboards
    * Visualize data in real-time and pre-defined schedule
    * Drag and drop products that do not require knowledge of progamming
### Conclusion:
* Automated tools, frameworks, and processes for all stages of the data analytics process are part of the Data Engineer's Ecosystem
* It is a diverse, rich, and challenging ecosystem
